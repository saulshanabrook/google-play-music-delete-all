```python
PW = ''
from gmusicapi import Mobileclient

api = Mobileclient()
logged_in = api.login('s.shanabrook@gmail.com', PW, Mobileclient.FROM_MAC_ADDRESS)
```

```python
songs = api.get_all_songs()
```

```python
songs
```

```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]

for c in chunks(songs, 100):
    print("deleting")
    api.delete_songs([s['id'] for s in c])
```
